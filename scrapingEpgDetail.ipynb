```
def getStartTime (soup):
   
  table_block = soup.find(class_="program_detail_table1_td")
  tr_tag = table_block.find_all("tr")
  # print(tr_tag)
  pattern4 = r"(\d+):\d+"

  start_time = None
  for tmp_tr in tr_tag:
    # print(tmp_tr)
    if re.findall(pattern4, str(tmp_tr)):

      start_time = re.findall(pattern4, str(tmp_tr))
  

  return(start_time)  
```

```
def checkStartTime (soup):
   
  table_block = soup.find(class_="program_detail_table1_td")
  tr_tag = table_block.find_all("tr")
  # print(tr_tag)
  pattern4 = r"(\d+):\d+"

  start_time = None
  for tmp_tr in tr_tag:
    # print(tmp_tr)
    if re.findall(pattern4, str(tmp_tr)):

      start_time = re.findall(pattern4, str(tmp_tr))
  
  if  int(str(start_time[0])) < 17:
    return(False)

  return(True)  
```

```
def getSharp (soup):
   
  # soup2 = BeautifulSoup(html, fromEncoding="utf-8")
  table_block = soup.find(class_="program_detail_table1_td")
  tr_tag = table_block.find_all("tr")
  # print(tr_tag)
  pattern3 = r"#\d+"

  sharp_no = None
  for tmp_tr in tr_tag:
    # print(tmp_tr)
    if re.findall(pattern3, str(tmp_tr)):

      sharp_no = re.findall(pattern3, str(tmp_tr))
  

  return(sharp_no)  


```

```
def getSharpTitle (soup):
   
  # soup2 = BeautifulSoup(html, fromEncoding="utf-8")
  table_block = soup.find(class_="program_detail_table1_td")
  tr_tag = table_block.find_all("tr")
  # print(tr_tag)
  pattern6 = r".+番組名.+"

  sharp_title = None
  for tmp_tr in tr_tag:
    # print(tmp_tr)
    if re.findall(pattern6, str(tmp_tr)):
      # breakpoint()a
      sharp_title= tmp_tr.text
  

  return(sharp_title)  


```

```
def getDetailOfProgram (soup) :
  # soup2 = BeautifulSoup(html, fromEncoding="utf-8")
  table_block = soup.find(class_="program_detail_table1_td")
  tr_tag = table_block.find_all("tr")
  # print(tr_tag)
  pattern2 = r".+番組詳細.+"

  detail_p = None
  for tmp_tr in tr_tag:
    if re.findall(pattern2, str(tmp_tr)):
      # breakpoint()a
      detail_p = tmp_tr.text
  

  return(detail_p)  

```

```
def checkTime (soup):
   
  # soup = BeautifulSoup(html, fromEncoding="utf-8")
  b_waku = soup.find_all(class_="program_detail_data")

  pattern = r"（(\d+)分）"

  for tmp_b_waku in b_waku:
    b_time = re.findall(pattern, str(b_waku[1]))
  if not b_time :
    return(True)
  if int( str(b_time[0]) ) < 20:
    return ( False )
  return( True )
```

```
error_cnt = 0

def openUrl (url):
  soup = None
  global error_cnt
  try:
    html = request.urlopen(url)
    # time.sleep(1)
    soup = BeautifulSoup(html, fromEncoding="utf-8")
  except urllib.error.URLError as e:
    print(e)
    error_cnt +=1

  return( soup)
```
