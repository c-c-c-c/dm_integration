```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
# Suppress warnings 
import warnings
warnings.filterwarnings('ignore')
import gc

# matplotlib and seaborn for plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#追記
import json
import datetime
import math
import re


#importじゃないけど、個人的には省略されて良いことがあまりないので・・
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 1000)

```

```python
broadcaster_dic = {
    "日本テレビ" : "NTV",	
    "読売テレビ" : "NTV",	
    "TBS" : "TBS",	
    "フジテレビ" : "CX",	
    "関西テレビ" : "CX",	
    "テレビ朝日" : "EX",	
}
```

```python
df_vr = pd.read_csv("../data/viewing_rating.csv", encoding='cp932')
```

```python
# df_cat = pd.read_excel("../data/drama_category_list.xlsx") 
# df_cat = pd.read_csv("../data/drama_category.csv")
import codecs
with codecs.open("../data/drama_category3.csv", "r", "Shift-JIS", "ignore") as file:
    df_cat = pd.read_table(file, delimiter=",")
```

```python
################
# 視聴率csvからのdrama_key作成
### 
df_vr.start_at = pd.to_datetime(df_vr.start_at)
df_vr["drama_key"] = df_vr.start_at.dt.strftime("%y")+df_vr.start_at.dt.strftime("%m")+"_"+df_vr.TV_station+"_"+df_vr.time_table
```

```python
def regexTime (tmp_str) :
#     m = re.match(r"曜\s?(\d+):", tmp_str)
    m = re.findall(r"曜\s?(\d+):", tmp_str)
    if not m :
        return ("other")
    return m[0]
```

```python
##############
#policeとlove関連のカテゴリーをまとめる
#
police_cat_conf = ["刑事ドラマ","サスペンスドラマ","推理ドラマ","ミステリードラマ", "犯罪ドラマ",
                  "弁護士ドラマ","ミステリー","検事ドラマ","警察ドラマ","推理コメディ",
                   "刑事コメディー"]

love_cat_conf = ["恋愛ドラマ","ラブストーリー" ,"恋愛コメディ" , "ラブコメディー"]
```

```python
for i in range(len(df_cat)):
    if df_cat["drama_category"].iloc[i] in police_cat_conf:
        df_cat["drama_category_grouped"].iloc[i] = "刑事ドラマ"
    elif df_cat["drama_category"].iloc[i] in love_cat_conf:
        df_cat["drama_category_grouped"].iloc[i] = "恋愛ドラマ"
    else:
        df_cat["drama_category_grouped"].iloc[i] = "その他のドラマ"
    
```

```python
#########
# df_catにドラマキーを追加
df_cat["drama_key"] = \
    df_cat["start_at_year"].apply(lambda x: str(x)[2:4]) + \
    df_cat["start_at_month"].apply(lambda x: str(x)[1:3])\
    +"_"+ df_cat["broadcaster"].apply(lambda x : broadcaster_dic.get(x)) +"_"+\
    df_cat["weekday_time"].apply(lambda x : str(x)[0])+\
    df_cat["weekday_time"].apply(lambda x : regexTime(str(x) ) )
```

```python
no_drama_key = []
police_keys = []
love_keys = []

for tmp_key in df_vr.drama_key:
#     print(tmp_key)
    if tmp_key in df_cat["drama_key"].values:
        print (tmp_key +" exists"+ str(df_cat[df_cat["drama_key"] == tmp_key].drama_category.get_values()[0]) )
        print (tmp_key +" exists"+ str(df_cat[df_cat["drama_key"] == tmp_key].drama_category.get_values()[0]) )
#         if str(df_cat[df_cat["drama_key"] == tmp_key].drama_category.get_values()[0]) == "刑事ドラマ" :
        if str(df_cat[df_cat["drama_key"] == tmp_key].drama_category_grouped.get_values()[0]) == "刑事ドラマ" :
            police_keys.append(tmp_key)
#         if str(df_cat[df_cat["drama_key"] == tmp_key].drama_category.get_values()[0]) == "恋愛ドラマ" :
        if str(df_cat[df_cat["drama_key"] == tmp_key].drama_category_grouped.get_values()[0]) == "恋愛ドラマ" :
            love_keys.append(tmp_key)
    else:
        print("____NO "+tmp_key + df_vr[df_vr["drama_key"]==tmp_key].drama_title.values[0] )
        no_drama_key.append(tmp_key)
# df_cat["drama_key"]
```
