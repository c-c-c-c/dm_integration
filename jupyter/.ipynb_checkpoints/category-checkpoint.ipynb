```python
broadcaster_dic = {
    "日本テレビ" : "NTV",	
    "読売テレビ" : "NTV",	
    "TBS" : "TBS",	
    "フジテレビ" : "CX",	
    "関西テレビ" : "CX",	
    "テレビ朝日" : "EX",	
}
```

```python
df_vr = pd.read_csv("../data/viewing_rating.csv", encoding='cp932')
```

```python
# df_cat = pd.read_excel("../data/drama_category_list.xlsx") 
# df_cat = pd.read_csv("../data/drama_category.csv")
import codecs
with codecs.open("../data/drama_category3.csv", "r", "Shift-JIS", "ignore") as file:
    df_cat = pd.read_table(file, delimiter=",")
```

```python
################
# 視聴率csvからのdrama_key作成
df_vr.start_at = pd.to_datetime(df_vr.start_at)
df_vr["drama_key"] = df_vr.start_at.dt.strftime("%y")+df_vr.start_at.dt.strftime("%m")+"_"+df_vr.TV_station+"_"+df_vr.time_table
```

```python
def regexTime (tmp_str) :
#     m = re.match(r"曜\s?(\d+):", tmp_str)
    m = re.findall(r"曜\s?(\d+):", tmp_str)
    if not m :
        return ("other")
    return m[0]
```

```python
regexTime("月曜 21:00")
```

```python
df_cat["drama_key"] = \
    df_cat["start_at_year"].apply(lambda x: str(x)[2:4]) + \
    df_cat["start_at_month"].apply(lambda x: str(x)[1:3])\
    +"_"+ df_cat["broadcaster"].apply(lambda x : broadcaster_dic.get(x)) +"_"+\
    df_cat["weekday_time"].apply(lambda x : str(x)[0])+\
    df_cat["weekday_time"].apply(lambda x : regexTime(str(x) ) )
# df_cat["weekday_time"].apply(lambda x : re.match(r"月曜",str(x))  )
# df_cat["weekday_time"].apply(lambda x : str(x)  )
# df_cat.columns
```

```python
no_drama_key = []
for tmp_key in df_vr.drama_key:
#     print(tmp_key)
    if tmp_key in df_cat["drama_key"].values:
        print (tmp_key +" exists")
    else:
        print("____NO "+tmp_key + df_vr[df_vr["drama_key"]==tmp_key].drama_title.values[0] )
        no_drama_key.append(tmp_key)
# df_cat["drama_key"]
```

```python

for tmp_no_key in no_drama_key:
    print(tmp_no_key + "_" +  df_vr[df_vr["drama_key"] ==tmp_no_key].drama_title.values[0] )
```
