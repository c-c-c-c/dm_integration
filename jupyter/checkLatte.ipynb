```python
# time型に
df_vr.start_at=pd.to_datetime(df_vr.start_at)
# averageの％を削除
# vr["average"] =  vr["average"].apply(lambda x : float(x[:-1]) )
```

```python
df_vr["drama_key"] = df_vr.start_at.dt.strftime("%y")+df_vr.start_at.dt.strftime("%m")+"_"+df_vr.TV_station+"_"+df_vr.time_table
```

```python
no_tv = []
for d_key in df_vr.drama_key:
    if d_key in df["drama_key"].values:
        print(d_key+" exists!!")
        pass
    else:
        print("_______"+d_key+"  nowhere" )
        no_tv.append(d_key)
        
```

```python
# df[df.drama_key =='1907_CX_月21'
no_tv_dict = {}
for d_key in no_tv:
    
    df_vr[df_vr["drama_key"] == d_key].drama_title
    no_tv_dict[d_key] = df_vr[df_vr["drama_key"] == d_key].drama_title.values[0]
    #     a["drama_title"]
```

```python
df_no_tv_dict=  pd.DataFrame([no_tv_dict.keys(),list(no_tv_dict.values())])
```
