```python
regex = r"血液型|事務所|生年月日|出身地|出身|最終学歴|活動期間|職業|ジャンル"
```

```python
from urllib import request
from bs4 import BeautifulSoup
import urllib.request
import requests
import re
from urllib.error import HTTPError
```

```python
# URLごとにスクレイピングを行う
key_list = []
error_actor = []

def scraping(url, actor_name):
    tmp_actor_profile_dict = {actor_name : {}}
    #ページに複数ある対象部分を取得する
    try:
        html = request.urlopen(url)
    except HTTPError as e:
        error_actor.append(actor_name)
        return(tmp_actor_profile_dict)
        
    soup = BeautifulSoup(html, fromEncoding="utf-8")
    actor_info_tables = soup.select('.infobox tr')
    
    for i, dt in enumerate(actor_info_tables):
#         breakpoint()
        print(type(dt.contents))
        if type(dt.contents) is list and len(dt.contents) > 1:
#             and type(dt.contents[0]) is NavigableString:
            
            try:
                search_result = re.search(regex,str(dt.contents[0].text))
            
                if search_result is not None and dt.contents[0].text is not None:
                    print("key:"+str(dt.contents[0].text))
                    print("val:"+str(dt.contents[1].text))

                    tmp_actor_profile_dict[actor_name][dt.contents[0].text] =\
                    dt.contents[1].text
            except AttributeError as e:
                error_actor.append(actor_name)
                
                
    return tmp_actor_profile_dict
```

```python
import json

f = open("../data/drama_info0115.json", 'r')
print(f)
drama_info_json = json.load(f)
```

```python
actor_actress_list =[]

for tmp_key in drama_info_json:
    
    actor_actress_list.append(drama_info_json[tmp_key].get("main_actor1"))
    actor_actress_list.append(drama_info_json[tmp_key].get("main_actor2"))
```

```python
actor_actress_list = list(set(actor_actress_list))
```

```python
import urllib.parse
###########
# 実行本体
###
actor_profile_dict = {}
import pprint
# pprint.pprint(actor_profile_dict)

for tmp_name in actor_actress_list:
    actor_name = urllib.parse.quote(tmp_name)
    tmp_actor_profile_dict = scraping("https://ja.wikipedia.org/wiki/"+ actor_name,tmp_name)

    actor_profile_dict.update(tmp_actor_profile_dict)
    
```
