```python
from urllib import request
from bs4 import BeautifulSoup
import urllib.request
import requests
import re
from urllib.error import HTTPError
```

```python
# URLごとにスクレイピングを行う
key_list = []
error_actor = []

def scraping(url, actor_name):
    tmp_actor_profile_dict = {actor_name : {}}
    #ページに複数ある対象部分を取得する
    try:
        html = request.urlopen(url)
    except HTTPError as e:
        error_actor.append(actor_name)
        return(tmp_actor_profile_dict)
        
    soup = BeautifulSoup(html, fromEncoding="utf-8")
    actor_info_tables = soup.select('.infobox tr')
    
    for i, dt in enumerate(actor_info_tables):
#         breakpoint()
        print(type(dt.contents))
        if type(dt.contents) is list and len(dt.contents) > 1:
#             and type(dt.contents[0]) is NavigableString:
            
            try:
                search_result = re.search(regex,str(dt.contents[0].text))
            
                if search_result is not None and dt.contents[0].text is not None:
                    print("key:"+str(dt.contents[0].text))
                    print("val:"+str(dt.contents[1].text))

                    tmp_actor_profile_dict[actor_name][dt.contents[0].text] =\
                    dt.contents[1].text
            except AttributeError as e:
                error_actor.append(actor_name)
                
                
    return tmp_actor_profile_dict
```

```python
actor_profile_dict_shaped = {}
for tmp_actor in actor_profile_dict.keys():
    actor_profile_dict_shaped[tmp_actor] = {} 
    tmp_profile = actor_profile_dict[tmp_actor]
    for tmp_key in tmp_profile.keys():
        if re.match( r"生年月日" ,tmp_key ):
            b_day_char = tmp_profile[tmp_key]
            tmp_match = re.findall(r"(\d+-\d+-\d+)" , b_day_char)
            if tmp_match:
                print(tmp_match[0])
                actor_profile_dict_shaped[tmp_actor]["birthday"] = tmp_match[0]
        elif re.match( r"事務所" ,tmp_key ):
            office_char = tmp_profile[tmp_key]
            office_char = re.sub(r"\n" , "", office_char)
            actor_profile_dict_shaped[tmp_actor]["office"] = office_char
        elif re.match( r"血液型" , tmp_key):
            blood_type = tmp_profile[tmp_key]
            blood_type = re.sub(r"\n" , "", blood_type)
            blood_type = re.sub(r"型" , "", blood_type)
            blood_match = re.match(r"(A|B|O|AB)" ,blood_type)
#             print(blood_type2[0])
            if blood_match : 
                actor_profile_dict_shaped[tmp_actor]["blood_type"] = blood_match[0]
            
    
```
